```python
def bar_Chart(x,y,name,yaxis,marker_line_color,textfontcolor):
    chart = go.Bar(
        x=x,
        y=y,
        name=name,
        yaxis=yaxis,
        text = y,
        textposition='outside',
        marker=dict(
        color=marker_line_color
    ),
        textfont=dict(size=13,
                      color=textfontcolor),      
                      marker_line_color=marker_line_color,
                      marker_line_width=1, 
                      opacity=0.8
                    )
    return chart

def Scatter_Chart(x,y,name,yaxis,markercolor,linecolor):
    chart = go.Scatter(
        x=x,
        y=y,
        name=name,
        yaxis=yaxis,
        mode='lines', 
        #marker= dict(size=3,
        #            symbol = 'diamond',
        #            color =markercolor,
        #            line_width = 2),
        line = dict(color=linecolor, width=2)
        )
    return chart

def produce_yaxis(title,txt_y,color):#藍 "#00008b" 紅 "#dc143c" 橘 "#d2691e"
    if txt_y==1:
        yaxis=dict(
            title=title,
            titlefont=dict(
                color=color
            ),
            tickfont=dict(
                color=color
            ),
            side="left",
        position=0.05
        )
    elif txt_y==2:
        yaxis=dict(
        title=title,
        titlefont=dict(
            color=color
        ),
        tickfont=dict(
            color=color
        ),
        anchor="x",
        overlaying="y",
        side="right",
        )
    else :
        yaxis=dict(
        title=title,
        titlefont=dict(
            color=color
        ),
        tickfont=dict(
            color=color
        ),
        anchor="free",
        overlaying="y",
        side="right",
        position=0.94
    )
    return yaxis

```
